# Más allá de la diversificación



---

## I. Los límites de la diversificación

La diversificación es una herramienta fundamental en la gestión de portafolios. Sus beneficios principales incluyen:

- **Reducir el riesgo idiosincrático**: Al incluir activos no perfectamente correlacionados, se eliminan riesgos específicos que no son compensados por el mercado.
- **Optimizar la rentabilidad ajustada al riesgo**: Permite construir una cartera eficiente, maximizando el retorno esperado por unidad de riesgo asumido.

Sin embargo, la diversificación tiene limitaciones importantes:

- Funciona **en promedio y en condiciones normales de mercado**, pero **puede fallar en momentos críticos**, justo cuando más se necesita.
- Por ejemplo, durante la **crisis financiera de 2008**, los activos que usualmente no estaban correlacionados comenzaron a moverse en la misma dirección, anulando los beneficios de la diversificación.
- Esto ocurre porque la diversificación **no protege contra el riesgo sistemático**, es decir, los movimientos del mercado en su conjunto.

### Alternativas a la Diversificación

Para mitigar los riesgos sistemáticos y enfrentar escenarios extremos, los inversionistas pueden recurrir a otras estrategias como la **cobertura** y los **seguros financieros**.

### 1. Cobertura (Hedging)

La cobertura busca proteger una cartera frente a riesgos sistemáticos mediante la inclusión de activos cuya exposición sea similar (y de signo opuesto) a la de los pasivos.

#### Características clave:

- Permite **limitar pérdidas ante caídas del mercado**, pero también puede **reducir ganancias potenciales**.
- Es más adecuada para **inversionistas con alta aversión al riesgo** o que buscan seguridad financiera.
- Puede implementarse utilizando **derivados financieros** como:
  - **Opciones Put**: Protegen contra caídas en el precio de una acción.
  - **Futuros y Forwards**: Se usan para fijar precios de activos y mitigar la volatilidad.

#### Ejemplo:
Un inversionista que posee acciones de una empresa puede comprar una opción put sobre esas acciones para limitar sus pérdidas si el precio cae.

### 2. Seguros Financieros (Insurance)

A diferencia de la cobertura estática, los seguros ofrecen una **protección más eficiente** al permitir cierta participación en las subidas del mercado mientras se limita la exposición a caídas.

#### Características clave:

- Se implementan a través de **cobertura dinámica**.
- Utilizan instrumentos financieros como **swaps** o **seguros tradicionales**.
- Son ampliamente usados por:
  - **Empresas**, para cubrir riesgos operativos y financieros.
  - **Fondos de pensiones**, para proteger los ahorros frente a la volatilidad de los mercados.

### 3. Cobertura Dinámica

La cobertura dinámica ajusta la exposición de la cartera de forma continua según las condiciones del mercado.

#### Ventajas:

- Combina **protección contra pérdidas** y **potencial de crecimiento**.
- Mejora la eficiencia de la gestión de riesgos.
- Se adapta a entornos cambiantes.

#### Estrategias comunes:

- **Gestión basada en opciones**: Se ajusta la posición en derivados dependiendo del comportamiento del mercado.
- **Paridad de riesgo (Risk Parity)**: Reequilibra los activos en función de su contribución al riesgo total, manteniendo un nivel constante de exposición.

---

## II. Constant Proportion Portfolio Insurance - CPPI

El CPPI (Seguro de Cartera de Proporción Constante) es una estrategia que permite al inversor **mantener la exposición al potencial alcista de un activo de riesgo, a la vez que ofrece una garantía contra el riesgo bajista** mediante el escalado dinámico de la exposición.

##### Procedimiento:

1. Imagine una cartera compuesta por dos tipos de activos: **"seguros"** y **"arriesgados"**.

2. Se define un **nivel de protección** (por ejemplo, el valor inicial del portafolio, s) y un **horizonte de inversión** (t).

3. Podemos calcular cuánto **puede disminuir el valor de la cartera en cada momento t**, de modo que el activo seguro pueda recuperar la pérdida hasta alcanzar el nivel de protección:

$$
F_t = \frac{s}{(1+y_t)^{(T-t)}}
$$

Donde, $F_t$ es el **piso (floor)**: el valor mínimo que se desea proteger, descontando el rendimiento del activo seguro a lo largo del tiempo. El **colchón (cushion)** es la diferencia entre el valor actual del portafolio y el piso. Esta es la parte que se puede arriesgar.

4. La estrategia es que si el valor del portafolio se acerca al piso, el cushion se reduce y, por tanto, también la exposición al riesgo.

![CCPI](data/CPPI.png)

##### Ventajas:

- Si se pudiera operar continuamente, **nada podría salir mal**.
- Crea **payoffs similares a una opción**, sin necesidad de comprar derivados.
- Permite exposición al crecimiento del mercado con una protección integrada.

##### Desventajas:

- **Gap Risk**: si ocurre una caída muy fuerte entre dos fechas de rebalanceo, el portafolio puede romper el piso. Esto ocurre si la pérdida excede: $ \frac{1}{M} $ en un intervalo de trading.

- **Costos de transacción**: a mayor frecuencia de rebalanceo, mayores costos.

---

## II.I Extensiones del Modelo

### 1. Extensión de Máximo Drawdown

Esta extensión modifica el CPPI tradicional estableciendo el **nivel de protección** como un porcentaje máximo de pérdida aceptable desde el **valor más alto alcanzado por el portafolio**. Es decir, en lugar de proteger solo el valor inicial, se protege un porcentaje del **máximo histórico** del portafolio.

#### Ejemplo:
Supongamos que el inversor desea limitar el drawdown máximo al **20%**. Entonces, el portafolio no debería caer por debajo del **80% del valor máximo histórico** alcanzado:

**Fórmula para el piso con drawdown máximo aceptable**:  
$$
V_t \geq \alpha * M_T
$$
Este piso dinámico se basa en el proceso de máximo histórico $ M_t $, conocido como **"running max"**, que en el tiempo $t$ se define como el mayor valor alcanzado por el portafolio entre el inicio y el tiempo actual.

Este proceso es **no decreciente**, ya que solo se mantiene igual o crece. Por lo tanto, el piso también es dinámico y se ajusta automáticamente con cada nuevo máximo alcanzado.

La asignación al activo riesgoso se calcula entonces como:

**Asignación al activo riesgoso**:  
$$
FM_t = \frac{\alpha * M_t}{(1+y_t)^{(T-t)}}
$$

> Si se fija una pérdida máxima aceptable de **0%** (es decir, $ \alpha = 100\% $, el modelo protege **cada nuevo máximo alcanzado** como si fuera el nuevo capital a proteger.

---

### 2. Extensión con Cap (Límite Superior)

Esta extensión introduce una **restricción superior (cap)** además del **piso (floor)**. El objetivo es limitar también el potencial alcista del portafolio para poder **reducir el riesgo de pérdida**.

La lógica es: si el inversor está dispuesto a renunciar a parte de las ganancias potenciales, puede protegerse mejor contra las pérdidas.

Se define un **umbral (threshold)** entre el piso y el cap, lo que divide el rango de valores posibles del portafolio en tres zonas:

- Zona de **riesgo**: entre el piso y el umbral.
- Zona de **transición**: alrededor del umbral.
- Zona de **limitación de ganancia**: entre el umbral y el cap.

**Límites inferior y superior del umbral**:  
$$
F_t \leq Q_t \leq CAP_t
$$

#### Asignación al Activo Riesgoso:

- Si el valor del portafolio está **entre el piso y el umbral**, se usa la fórmula tradicional del CPPI:

**Cushion clásico**:  
$$
F_t \leq P_t \leq Q_t
\qquad\Longrightarrow\qquad
E_t = M(P_t - F_t)
$$

- Si el valor del portafolio está **entre el umbral y el cap**, el modelo ajusta la asignación para evitar que el portafolio supere el cap:

**Cushion inverso**:  
$$
Q_t \leq P_t \leq CAP_t
\qquad\Longrightarrow\qquad
E_t = M(CAP_t - P_t)
$$

#### Determinación del Umbral

Para asegurar continuidad en la asignación al activo riesgoso al pasar de una zona a otra, se impone la condición de **"smooth-pasting"**. Esto implica que, en el punto umbral, ambas fórmulas de asignación deben coincidir:

**Condición de continuidad ("smooth-pasting")**:  
$$
 M(P_t - F_t) =  M(CAP_t - P_t)
$$

De esta condición se deduce una respuesta simple:

**Fórmula para el umbral**:  
$$
Q_t = \frac{F_t + CAP_t}{2}
$$

## Parte Práctica 


```python
# Librerías a utilizar
import pandas as pd
import numpy as np

# Gráficos
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import ticker

import seaborn as sns
import ipywidgets as widgets

# Archivo personalizado
import edhec_kit as erk
```

Vamos a crear una estrategia de inversión usando los retornos de industrias y los retornos del índice total de mercado.

```python
industry_r = erk.get_ind_file(filetype='returns') ## Importamos los datos de retornos mensuales por industria
market_index_r = erk.get_total_market_index_returns() ## Importamos los retornos del índice total del mercado
```

Consideraremos los retornos desde el 2000 para tres industrias. Además, creamos un conjunto de activos libre de riesgo con un retorno anual de 3%.

```python
risky_rets = industry_r['2000':][['Steel', 'Fin', 'Beer']]  ## Seleccionamos los retornos de tres industrias desde el año 2000

safe_rets    = pd.DataFrame().reindex_like(risky_rets)
safe_rets[:] = 0.03 / 12

risky_rets.head()
```

Ahora damos valor a nuestra cuenta incial $ A $, el piso (floor), el valor del piso y el multiplicar (m)

```python
valor_inicial = 1000
valor_cuenta = 1000
m = 3
piso = 0.8
valor_piso = piso * valor_cuenta
```

Creamos un data frame con el mismo índice que los activos riesgosos para nuestros valores creados anteriormente

```python
account_history = pd.DataFrame().reindex_like(risky_rets)
cushion_history = pd.DataFrame().reindex_like(risky_rets)
risky_w_history = pd.DataFrame().reindex_like(risky_rets)
```

Calculemos el crecimiento de la riqueza si solo hubiéramos invertido en activos de riesgo.

```python
risky_wealth = valor_inicial * (1 + risky_rets).cumprod()
risky_wealth.head()
```

A continuación, implementamos una estrategia CPPI, que ajusta dinámicamente la exposición a activos riesgosos en función de un "colchón" (cushion) entre el valor actual del portafolio y un piso mínimo garantizado (floor)

```python
# Iteramos sobre cada periodo de tiempo (por ejemplo, cada mes)
for paso in range(len(risky_rets.index)):
    
    # Calculamos el cushion: porcentaje del valor que está por encima del piso
    cushion = (valor_cuenta - valor_piso) / valor_cuenta

    # Calculamos la proporción a invertir en el activo riesgoso según el multiplicador m
    peso_riesgoso = m * cushion
    peso_riesgoso = np.clip(peso_riesgoso, 0, 1)  # Lo limitamos entre 0 y 1

    # Proporción restante va al activo libre de riesgo
    peso_seguro = 1 - peso_riesgoso

    # Asignamos el capital a cada tipo de activo
    inversion_riesgosa = peso_riesgoso * valor_cuenta
    inversion_segura   = peso_seguro   * valor_cuenta

    # Actualizamos el valor total del portafolio con los rendimientos del periodo
    valor_cuenta = (
        inversion_riesgosa * (1 + risky_rets.iloc[paso]) +
        inversion_segura   * (1 + safe_rets.iloc[paso])
    )

    # Guardamos los resultados para análisis posterior
    account_history.iloc[paso] = valor_cuenta
    cushion_history.iloc[paso] = cushion
    risky_w_history.iloc[paso] = peso_riesgoso

# Calculamos los retornos de la estrategia CPPI a partir del historial de cuenta
cppi_rets = (account_history / account_history.shift(1) - 1).dropna()
```

Graficamos la evolución de la riqueza del portafolio

```python
ax = account_history.plot(figsize=(10, 5), grid=True)
ax.set_ylabel('Riqueza del portafolio ($)')
ax.set_title('Evolución del portafolio con estrategia CPPI')
plt.show()
```

Ahora comparamos la evolución de la riqueza bajo una estrategia CPPI frente a una estrategia completamente expuesta al activo riesgoso (100%).

```python
# Creamos un gráfico con 3 filas y 2 columnas
fig, ax = plt.subplots(3, 2, figsize=(18, 15))
ax = ax.flatten()  # Aplanamos el array de ejes para poder indexar fácilmente

# Evolución de la riqueza: CPPI vs 100% en Beer
account_history['Beer'].plot(ax=ax[0], grid=True, label='CPPI en Beer')
risky_wealth['Beer'].plot(ax=ax[0], grid=True, label='100% en Beer', style='k:')
ax[0].axhline(y=valor_piso, color='r', linestyle='--', label='Piso fijo')
ax[0].legend(fontsize=11)
ax[0].set_title('Industria: Cerveza (Beer) - Riqueza')

# Evolución del peso del activo riesgoso
risky_w_history['Beer'].plot(ax=ax[1], grid=True, label='Peso riesgoso en Beer')
ax[1].legend(fontsize=11)
ax[1].set_title('Industria: Cerveza (Beer) - Peso Riesgoso')

# Evolución de la riqueza: CPPI vs 100% en Fin
account_history['Fin'].plot(ax=ax[2], grid=True, label='CPPI en Fin')
risky_wealth['Fin'].plot(ax=ax[2], grid=True, label='100% en Fin', style='k:')
ax[2].axhline(y=valor_piso, color='r', linestyle='--', label='Piso fijo')
ax[2].legend(fontsize=11)
ax[2].set_title('Industria: Financiera (Fin) - Riqueza')

# Evolución del peso del activo riesgoso
risky_w_history['Fin'].plot(ax=ax[3], grid=True, label='Peso riesgoso en Fin')
ax[3].legend(fontsize=11)
ax[3].set_title('Industria: Financiera (Fin) - Peso Riesgoso')

# Evolución de la riqueza: CPPI vs 100% en Steel
account_history['Steel'].plot(ax=ax[4], grid=True, label='CPPI en Steel')
risky_wealth['Steel'].plot(ax=ax[4], grid=True, label='100% en Steel', style='k:')
ax[4].axhline(y=valor_piso, color='r', linestyle='--', label='Piso fijo')
ax[4].legend(fontsize=11)
ax[4].set_title('Industria: Acero (Steel) - Riqueza')

# Evolución del peso del activo riesgoso
risky_w_history['Steel'].plot(ax=ax[5], grid=True, label='Peso riesgoso en Steel')
ax[5].legend(fontsize=11)
ax[5].set_title('Industria: Acero (Steel) - Peso Riesgoso')

# Mostrar el gráfico completo
plt.tight_layout()
plt.show()
```

Hasta ahora, la estrategia CPPI que implementamos mantenía un piso fijo en el valor de la cartera. Sin embargo, en algunos casos se prefiere que ese piso se ajuste dinámicamente, evitando caídas demasiado pronunciadas desde el punto más alto alcanzado por la riqueza.

Este mecanismo se conoce como restricción de reducción (drawdown constraint), y consiste en actualizar el piso en función del máximo histórico de la riqueza. Cuando esto se activa, el piso se vuelve móvil y se ajusta a:

$$ Piso dinámico=(1−reducciónn)×max(riqueza) $$

En caso de que exista una reducción, el método impone el multiplicador $m=1/\text{reducción}$.

Por ejemplo, consideremos $\text{reducción}=0,2$ (es decir, $20\%$):

```python
# Ejecutamos la estrategia CPPI con restricción de reducción del 20%
resultado = erk.run_cppi(
    risky_rets, 
    m=1/0.2,                  # multiplicador ajustado según la reducción
    start=1000,               # valor inicial de la cartera
    floor=0.8,                # piso inicial (usado solo si no hay drawdown)
    drawdown=0.2,             # restricción: pérdida máxima del 20%
    riskfree_rate=0.03        # tasa libre de riesgo anual
)

# Lista de industrias a graficar
industrias = ['Fin', 'Beer', 'Steel']

# Crear una figura con 3 filas y 2 columnas
fig, ax = plt.subplots(3, 2, figsize=(18, 12))
ax = ax.flatten()

# Iteramos sobre cada industria y su posición correspondiente en los subplots
for i, industria in enumerate(industrias):
    # Gráfico de riqueza acumulada
    resultado['Wealth'][industria].plot(
        ax=ax[2*i], 
        grid=True, 
        label=f'CPPI con Drawdown en {industria}'
    )
    resultado['Risky Wealth'][industria].plot(
        ax=ax[2*i], 
        grid=True, 
        label=f'100% en {industria}', 
        style='k:'
    )
    resultado['floor'][industria].plot(
        ax=ax[2*i], 
        color='r', 
        linestyle='--', 
        label='Piso dinámico'
    )
    ax[2*i].legend(fontsize=11)
    ax[2*i].set_title(f'Evolución de Riqueza - {industria}')

    # Gráfico de asignación al activo riesgoso
    resultado['Risky Allocation'][industria].plot(
        ax=ax[2*i + 1], 
        grid=True, 
        label=f'Asignación Riesgosa en {industria}'
    )
    ax[2*i + 1].legend(fontsize=11)
    ax[2*i + 1].set_title(f'Peso Riesgoso - {industria}')

# Ajustamos los espacios del gráfico
plt.tight_layout()
plt.show()
```

Observe ahora el comportamiento de la estrategia CPPI al variar la restricción de reducción máxima (drawdown)

```python
# Lista de sectores o industrias a analizar
industrias = ['Beer', 'Fin', 'Steel']

# Lista de restricciones de reducción máxima (drawdowns)
drawdowns = [0.2, 0.4, 0.6]

# Creamos la figura con dos subgráficos para cada industria
fig, ax = plt.subplots(3, 2, figsize=(18, 15))  # 3 filas, 2 columnas
ax = ax.flatten()

# Iteramos sobre cada industria
for i, industria in enumerate(industrias):
    # Graficamos la estrategia 100% en el activo riesgoso como referencia
    resultado_base = erk.run_cppi(
        risky_rets,
        m=1, start=1000, floor=0.8,
        drawdown=None, riskfree_rate=0.03
    )
    resultado_base['Risky Wealth'][industria].plot(
        ax=ax[2*i],
        grid=True,
        style='k:',
        label=f'{industria} (100% en riesgoso)'
    )

    # Creamos un DataFrame para almacenar las estadísticas
    resumen_estadisticas = pd.DataFrame()

    # Iteramos sobre cada nivel de drawdown
    for drawdown in drawdowns:
        resultado = erk.run_cppi(
            risky_rets,
            m=1/drawdown,
            start=1000,
            floor=0.8,
            drawdown=drawdown,
            riskfree_rate=0.03
        )

        # Graficamos la evolución de la riqueza para la industria
        resultado['Wealth'][industria].plot(
            ax=ax[2*i],
            grid=True,
            label=f'CPPI con dd={int(drawdown*100)}%, m={round(resultado["m"], 1)}'
        )

        # Graficamos el peso asignado al activo riesgoso para la industria
        resultado['Risky Allocation'][industria].plot(
            ax=ax[2*i+1],
            grid=True,
            label=f'dd={int(drawdown*100)}%, m={round(resultado["m"], 1)}'
        )

        # Guardamos las estadísticas de retorno de cada estrategia
        resumen = erk.summary_stats(
            resultado['Wealth'][[industria]].pct_change().dropna()
        )
        resumen_estadisticas = pd.concat([resumen_estadisticas, resumen], axis=0)

    # Títulos y leyendas
    ax[2*i].set_title(f'Evolución de la Riqueza - {industria}', fontsize=12)
    ax[2*i+1].set_title(f'Peso Asignado al Activo Riesgoso - {industria}', fontsize=12)
    ax[2*i].legend(fontsize=10)
    ax[2*i+1].legend(fontsize=10)

# Ajuste de layout para evitar superposición de gráficos
plt.tight_layout()
plt.show()
```
