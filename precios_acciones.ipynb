```python
import yfinance as yf
import pandas as pd
import datetime
```

```python
fecha_inicio = datetime.datetime(2019, 1, 1)
fecha_actual = datetime.datetime.now()

tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA', 'META', 'NVDA', 'JPM', 'V', 'JNJ', 'WMT', 'PG', 'DIS', 'MA', 'HD', 'BAC', 'XOM', 'NFLX', 'ADBE', 'PYPL', 'INTC', '^TNX']

data = yf.download(tickers, start=fecha_inicio, end=fecha_actual, auto_adjust=True)['Close']

data.head()
```

```python
data.to_excel('precios_acciones.xlsx')
```
